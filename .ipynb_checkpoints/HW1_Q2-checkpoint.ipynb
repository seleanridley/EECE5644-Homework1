```python
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from scipy.stats import multivariate_normal
import random
import pickle
from numpy import linalg as LA
import pylab as pl
```

```python
np.random.seed()

N = 1000;
samples = []
labels = []
priors = [0.3, 0.7]

#Class 1
size_1 = 0
mean_1 = np.array([-4, 2])
cov_1 = np.array([[2, 0], [0, 3]])

#Class 2
size_2 = 0;
mean_2 = np.array([3, 4])
cov_2 = np.array([[1.8, 0.3], [0.3, 1]])
```

```python
#Generate Samples
for i in range(N):
    u = random.random()
    if u <= priors[0]:
        labels.append(0)
        size_1 += 1
    else:
        labels.append(1)
        size_2 += 1


x1 = np.random.multivariate_normal(mean_1, cov_1, size_1)
x2 = np.random.multivariate_normal(mean_2, cov_2, size_2)

iter1 = iter(x1)
iter2 = iter(x2)
for k in range(len(labels)):
    if labels[k] == 0:
        n_value = next(iter1)
        samples.append( np.append( n_value, labels[k]) )
    else:
        n_value = next(iter2)
        samples.append( np.append( n_value, labels[k]) )
        
samples = np.array(samples).T
np.shape(samples)
```

```python
"""
    Minimum Expected Risk Classification
"""

#Likelihood functions

pdf1 = multivariate_normal.pdf(samples[:2].T, mean=mean_1, cov=cov_1)
pdf2 = multivariate_normal.pdf(samples[:2].T, mean=mean_2, cov=cov_2)


x = samples[:2]
decisions = []
class1, class2 = 0, 0

#loss = np.array([[0, 1], [1, 0]])
#gamma =  ( (loss[1, 1] - loss[0, 1])/ (loss[0, 0]-loss[1, 0] ) ) * ( priors[0]/priors[1] )

discrim_1 = []
discrim_2 = []

for i in range(N):
    if (pdf1[i] > pdf2[i]):
        decisions.append(0) #Choose class 1
        class1 += 1
    else:
        decisions.append(1)
        class2 += 1

decisions = np.array(decisions)
```

```python
X = np.arange(-10, 10, 0.1)
Y = np.arange(-10, 10, 0.1)
XY = np.meshgrid(X, Y)
#X = np.mgrid[-2:2:100j, -2:2:100j]
#points = np.c_(X.ravel(), Y.ravel())
print(np.shape(XY))
#print(XY[:10])

g1_x = []
g2_x = []
probs = [(class1/1000), (class2/1000)]
for i in range(1):
    for j in X:
        for k in Y:
            if i == 0:
                cov = cov_1
                mean = np.array([mean_1])
            else:
                cov = cov_2
                mean = np.array([mean_2])

            #x = np.array([points[:,t]])
            x = np.array([[k, j]])
            #print(x)
            W_i = -0.5 * LA.inv(cov)
            w_i = np.dot(LA.inv(cov), mean.T)
            w_i0 = (-0.5 * np.sum(np.dot(mean[0], LA.inv(cov)) * mean, axis=1 ))  + \
                        (-0.5 * np.log( np.linalg.det(cov) )) + \
                            (np.log(probs[i]))
            if i == 0:
                g1_x.append( [np.dot(np.dot(x[0], W_i), x.T) + np.dot(x[0], w_i) + w_i0][0][0])
            else:
                g2_x.append( [np.dot(np.dot(x[0], W_i), x.T) + np.dot(x[0], w_i) + w_i0][0][0])
                
g_x = np.array(g_x)
g_x.shape = len(X), len(Y)
```

```python
plt.figure(figsize=(20,10))
plt.rcParams.update({'font.size': 20})
for i in range(N):
    if samples[2, i] == 0:
        plt.plot(samples[0, i], samples[1, i], 'r.')
    else:
        plt.plot(samples[0, i], samples[1, i], 'b.')

plt.contourf(X, Y, g_x)
#plt.plot(XY, g_x)

#plt.contour(XY, discrim_1)

#plt.contour(XY, discrim_2)

#plt.show()
```
